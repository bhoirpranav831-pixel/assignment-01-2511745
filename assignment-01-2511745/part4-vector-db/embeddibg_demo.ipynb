```python
from sentence_transformers import SentenceTransformer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = [
    # Cricket
    "The batsman scored a century in the match",
    "The bowler took five wickets in the game",
    "Cricket is a popular sport in India",

    # Cooking
    "Add salt and spices while cooking the curry",
    "Baking a cake requires precise measurements",
    "The chef prepared a delicious meal",

    # Cybersecurity
    "Cybersecurity protects systems from attacks",
    "Use strong passwords to secure your accounts",
    "Hackers try to exploit system vulnerabilities",
    "Firewalls help in preventing unauthorized access"
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
similarity_matrix = cosine_similarity(embeddings, embeddings)

plt.figure(figsize=(10,8))
sns.heatmap(similarity_matrix, annot=True, cmap="coolwarm")
plt.title("Cosine Similarity Matrix")
plt.show()
```

```python
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 most similar sentences
top_indices = similarities.argsort()[-2:][::-1]

print("Query:", query)
print("\nTop 2 similar sentences:")

for i in top_indices:
    print(f"\nSentence: {sentences[i]}")
    print(f"Similarity Score: {similarities[i]:.4f}")
```
